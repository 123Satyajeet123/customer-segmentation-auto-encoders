```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('customer_offer_dataset.csv', index_col=0)
```

```python
offer_completed_columns = df.columns[28:]
offer_completed_columns
```

```python
def multi_hot(row):
    values = row.values
    # Find the indices that would sort the array
    sorted_indices = np.argsort(values)
    # We are interested in the top 3 values; argsort returns indices for ascending order, so we take the last three
    top_indices = sorted_indices[-3:]
    # Create a new row as pandas series
    new_row = pd.Series(np.zeros(len(values)), index=row.index)
    for i in top_indices:
        if values[i] > 0:
            new_row[i] = 1

    return new_row
```

```python
df.head()[offer_completed_columns]
```

```python
multi_hot_df = df.copy()
multi_hot_df[offer_completed_columns] = multi_hot_df[offer_completed_columns].apply(multi_hot, axis=1)

multi_hot_df.head()
```

```python
multi_hot_df.to_csv('customer_offer_dataset_multi_hot.csv')
```
